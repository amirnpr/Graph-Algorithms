# Maximum Independent Set of Tree

```python
import networkx as nx
import matplotlib.pyplot as plt

# Node class represents a node in the tree
class Node:
    def __init__(self, data):
        self.data = data
        self.children = []

# Builds a tree from the given arcs
def build_tree(arcs):
    nodes = {}
    for arc in arcs:
        u, v = arc
        # Create new node if not already present
        if u not in nodes:
            nodes[u] = Node(u)
        if v not in nodes:
            nodes[v] = Node(v)
        # Add child nodes to parent node
        nodes[u].children.append(nodes[v])
    return nodes

# Finds the maximum independent set of nodes in the tree
def maximum_independent_set(root):
    if not root:
        return set()
    
    # Case 1: Including the current node and excluding its children
    included = set([root.data])
    for child in root.children:
        for grandchild in child.children:
            included.update(maximum_independent_set(grandchild))

    # Case 2: Excluding the current node and including its children
    excluded = set()
    for child in root.children:
        excluded.update(maximum_independent_set(child))
    
    # Return the larger set of included and excluded nodes
    return included if len(included) > len(excluded) else excluded
```

# Example 1

```python
arcs = [('a', 'b'),
                 ('b', 'd'), 
        ('a', 'c'), 
                 ('c', 'e'), ('c', 'f')
        ]
tree = build_tree(arcs)
root_node = tree['a']
max_independent_set = maximum_independent_set(root_node)
print("Maximum Independent Set:", max_independent_set)

# Create a directed graph from the tree
G = nx.DiGraph()
for node in tree.values():
    G.add_node(node.data)
    for child in node.children:
        G.add_edge(node.data, child.data)

# Draw the tree
pos = nx.spring_layout(G)
plt.figure(figsize=(8, 4))
# Draw the nodes and edges of the graph
nx.draw_networkx(G, pos, with_labels=True, node_color='lightblue')
# Highlight nodes in the maximum independent set
nx.draw_networkx_nodes(G, pos, nodelist=list(max_independent_set), node_color='red')
plt.title("Visualization of Maximum Independent Set")
plt.show()
```
