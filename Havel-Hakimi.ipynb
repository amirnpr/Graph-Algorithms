# Havel - Hakimi

```python
def havel_hakimi(sequence):
    
    sequence = [x for x in sequence if x != 0]
    # If the sequence is empty, return True
    if not sequence:
        return True
    if any(x < 0 or x >= len(sequence) for x in sequence):
        return False

    # Descending Order
    sequence.sort(reverse=True)

    # Remove the first element 
    first = sequence.pop(0)

    # Return false if first element is greater than the length of the remaining sequence
    if first > len(sequence):
        return False
    # Subtract 1 
    for i in range(first):
        sequence[i] -= 1
        
    return havel_hakimi(sequence)
```

# Construct a graph from given degrees

```python
import matplotlib.pyplot as plt
import networkx as nx

def plotgraph (seq):

    G = nx.configuration_model(seq)

    degree_sequence = [d for n, d in G.degree()]
    hist = {}
    for d in degree_sequence:
        if d in hist:
            hist[d] += 1
        else:
            hist[d] = 1
    pos = nx.spring_layout(G)
    nx.draw(G)
    plt.show()
```

# Examples

```python
a = [4, 3, 2, 1, 1]
b = [3,3,3,1]
c = [7, 6, 4, 3, 3, 3, 2, 2]
d = [6,6,5,3,3,3,1]

print("a = ", havel_hakimi(a))  
print("b = ", havel_hakimi(b))  
print("c = ", havel_hakimi(c)) 
print("d = ", havel_hakimi(d)) 
```
